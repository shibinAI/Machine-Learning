Create a pipeline of the PCA() as transformer and Lasso as an estimator.
Use GridSearchCV for tuning the hyperparameters of the created pipeline on training dataset.
	Values of n_components for PCA to be [0.9, 0.95]
	lasso alpha value to be taken as : [10, 1, 0.01, 0.001]
	scoring : neg_mean_absolute_error.
	Use 5 fold cross validation.
	n_jobs = -1 (negative one) [it helps in using all the computational power to run this job]
(Note: Kindly ignore the warning.)

If we fit the pipeline on the training dataset, what will be the R2 score on the test dataset?

```python
from sklearn.pipeline import Pipeline
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
from sklearn.decomposition import PCA
```

```python
# Create a pipeline with PCA as transformer and Lasso as estimator
pipeline = Pipeline([
    ('pca', PCA()),
    ('lasso', Lasso(random_state=42))
])

# Define the hyperparameter grid for tuning
param_grid = {
    'pca__n_components': [0.9, 0.95],
    'lasso__alpha': [10, 1, 0.01, 0.001]
}

# Set up GridSearchCV
grid_search = GridSearchCV(
    estimator=pipeline,
    param_grid=param_grid,
    cv = 5,
    scoring = 'neg_mean_absolute_error',
    n_jobs=-1
)
# Fit on the training data
grid_search.fit(X_train, y_train)
```

```python
# Use the best estimator from the grid search to predict on the test set
best_pipeline = grid_search.best_estimator_
y_pred = best_pipeline.predict(X_test)
```

```python
# Calculate the r2 score
r2 = r2_score(y_test, y_pred)
```

```python
r2
```

How much variance is explained (Eigen value) by the first principle component?

```python
# Extract the fitted PCA object from the pipeline
pca = best_pipeline.named_steps['pca']

# Get the explained variance ratio array
explained_variance_ratio = pca.explained_variance_ratio_

# The variance explained by the fist principal component:
first_pc_variance = explained_variance_ratio[0]
print("Variance explained by the first principal component:", first_pc_variance)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.decomposition import PCA
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV

# Create a pipeline with PCA as transformer and Lasso as estimator
pipeline = Pipeline([
    ('pca', PCA()),
    ('lasso', Lasso(random_state=42))
])

# Define the hyperparameter grid
param_grid = {
    'pca__n_components': [0.9, 0.95],
    'lasso__alpha': [10, 1, 0.01, 0.001]
}

# Setup GridSearchCV with 5-fold CV, neg_mean_absolute_error as scoring, and use all CPUs
grid_search = GridSearchCV(estimator=pipeline,
                           param_grid=param_grid,
                           cv=5,
                           scoring='neg_mean_absolute_error',
                           n_jobs=-1)

# Fit GridSearchCV on the training data (assuming X_train and y_train are already defined)
grid_search.fit(X_train, y_train)

# Get the best pipeline from GridSearchCV
best_pipeline = grid_search.best_estimator_

# Extract the PCA step from the pipeline
pca = best_pipeline.named_steps['pca']

# Get the explained variance (eigenvalue) and explained variance ratio for the first principal component
first_pc_eigenvalue = pca.explained_variance_[0]
first_pc_explained_variance_ratio = pca.explained_variance_ratio_[0]

print("Eigenvalue (variance) explained by the first principal component:", first_pc_eigenvalue)
print("Explained variance ratio by the first principal component:", first_pc_explained_variance_ratio)
```
