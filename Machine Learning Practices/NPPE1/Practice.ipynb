```python
def get_category(x):
  if x < 10:
    return 'Category 1'
  elif 10 <= x < 20:
    return 'Category 2'
  elif 20 <= x < 30:
    return 'Category 3'
  elif 30 <= x < 40:
    return 'Category 4'
  else:
    return 'Category 5'
```

```python
df['CATEGORY'] = df['PRICE'].apply(get_category)
```

```python
df.head()
```

```python
conditions = [
    (df['PRICE'] < 10),
    ((df['PRICE'] >= 10) & (df['PRICE'] < 20)),
    ((df['PRICE'] >= 20) & (df['PRICE'] < 30)),
    ((df['PRICE'] >= 30) & (df['PRICE'] < 40)),
    (df['PRICE'] >= 40)
]
```

```python
choices = ['Category 1','Category 2','Category 3','Category 4','Category 5']
```

```python
import numpy as np
df['CATEGORY'] = np.select(conditions, choices, default = 'Unknown')
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

Drop the column 'CATEGORY'

```python
df = df.drop(columns = ['CATEGORY'])
df.head()
```

```python
transformer = ColumnTransformer(
    [
        ('mmscaler',MinMaxScaler(), ['CRIM','ZN','POLINDEX','DIS','HIGHWAYCOUNT','TAX','PTRATIO','IMM','BPL']),
        ('INDUS_stdscaler',StandardScaler(),['INDUS']),
        ('RM_pipeline', Pipeline([
            ('imputer', SimpleImputer(strategy='median')),
            ('mm_scaler', MinMaxScaler())
        ]),['RM']),
        ('AGE_pipeline',Pipeline([
            ('imputer', SimpleImputer(strategy='mean')),
            ('mm_scaler',MinMaxScaler())
        ]),['AGE']),
        ('RIVERSIDE', Pipeline([
            ('imputer', SimpleImputer(strategy='most_frequent')),
            ("encoder", OneHotEncoder())
        ]),["RIVERSIDE"])
    ],
    remainder="passthrough"
)
transformer
```

```python
X_train = transformer.fit_transform(X_train)
X_test = transformer.transform(X_test)
X_train.shape
```
