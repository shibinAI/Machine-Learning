```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/used_cars.csv')
df.head()
```

```python
df.isna().sum()
```

```python
df.shape
```

```python
df.describe()
```

```python
df = df.dropna()
df.shape
```

```python
X = df.drop(columns=['MSRP'])
```

```python
X.shape
```

```python
y = df['MSRP']
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
X_train.shape
```

```python
y_test.shape
```

BaggingRegressor(random_state=1)

RandomForestRegressor(random_state=1)

GradientBoostingRegressor(random_state=1)

AdaBoostRegressor(random_state=1)

```python
from sklearn.ensemble import BaggingRegressor, RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor
```

```python
# Initialize models
models = {
    "BaggingRegressor": BaggingRegressor(random_state=1),
    "RandomForestRegressor": RandomForestRegressor(random_state=1),
    "GradientBoostingRegressor": GradientBoostingRegressor(random_state=1),
    "AdaBoostRegressor": AdaBoostRegressor(random_state=1)
}
```

```python
from sklearn.metrics import r2_score
# Train and evaluate
for name, model in models.items():
  model.fit(X_train, y_train)
  y_pred = model.predict(X_test)
  r2 = r2_score(y_test, y_pred)
  print(f"{name} R² score:{r2}")
```

On the same data set, perform the preprocessing stated in the previous questions, and perform Hyperparameter Tuning using GridSearchCV on AdaBoostRegressor. Use random_state to be 1.

    Hyperparameter tuning to be done over:
        Use n_estimators as [100,500,1000]
        Use learning_rate as [0.5,1,2]

    Train the 'model' and compute the 'score' on the test data.

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'n_estimators':[100,500,1000],
              'learning_rate':[0.5,1,2]}

AdaBoostRegressor_model = AdaBoostRegressor(random_state=1)

grid_search = GridSearchCV(
    param_grid=param_grid,
    estimator=AdaBoostRegressor_model,
    cv=4,
    n_jobs=5
)
```

```python
grid_search.fit(X_train, y_train)
```

Enter the value of the 'score' on testing set.

Take cv value= 4

```python
best_model = grid_search.best_estimator_
```

```python
test_score = best_model.score(X_test, y_test)
print(test_score)
```
