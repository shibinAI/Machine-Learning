**Conditional Slicing**

```python
age_df_temp1 = df.loc[df.age < 5.383060e-03]
age_df_temp1
```

```python
age_df_temp2 = df.loc[df.age > 5.383060e-03]
age_df_temp2
```

```python
age_df_temp2.iloc[0]
```

loc will give an error if we give an index which is not actually present in the data

```python
age_df_temp1.loc[0]
```

iloc will not give error even if the exact index number is not present. It will just count and take

```python
age_df_temp1.iloc[0]
```

Conditional slicing can be done without even using loc

```python
age_df = df[df.age > 5.383060e-03]
age_df
```

**Using lambda function**

```python
selector = lambda df1: df1['A'] > 0.5
selector
```

```python
df1.loc[selector]
```

```python
selector1 = lambda df1: (df1['A'] > 0.5) & (df1['B'] < 0.2)
```

```python
df1.loc[selector1]
```

```python
condition_for_selection = (df1['A'] > 0.5) & (df1['B'] < 0.2)
condition_for_selection
```

```python
df1.loc[condition_for_selection]
```

```python
df1[condition_for_selection]
```

OR Operator |
NOT Operator ~

```python
condition_for_selection_OR = (df1['A'] > 0.5) | (df1['B'] < 0.2)
condition_for_selection_OR
```

```python
df1[condition_for_selection_OR]
```

```python
condition_for_selection_NOT = (df1['A'] > 0.5) | ~(df1['B'] < 0.2)
condition_for_selection_NOT
```
