Creating pandas dataframe from series

```python
cities = pd.Series(['Mumbai', 'Bangalore', 'Chennai','Delhi'])
population = pd.Series([17000000, 13000000, 6000000])
print(cities)
print(type(cities))
```

```python
city_info_df = pd.DataFrame({'City':cities, 'Population':population})
print(city_info_df)
```
