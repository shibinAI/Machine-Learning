```python
# Determine the six most recent years in the dataset
recent_years = sorted(df['Year'].unique(), reverse=True)[:6]

# Filter rows that belong to these years and count them
num_recent_rows = len(df[df['Year'].isin(recent_years)])
print(num_recent_rows)
```
