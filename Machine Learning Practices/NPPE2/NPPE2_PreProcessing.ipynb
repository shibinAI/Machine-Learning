Gender: Impute with most frequent then ordinally encode.

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder
gender_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy = 'most_frequent')),
    ('encoder', OrdinalEncoder())
])
```

Age: Impute with mean then standard scale.

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

age_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy = 'mean')),
    ('scaler', StandardScaler())
])
```

HasTension: Ordinally encode.

```python
from sklearn.preprocessing import OrdinalEncoder
tension_pipeline = Pipeline([
    ('encoder', OrdinalEncoder())
])
```

AnyHeartDisease:Ordinally encode.

```python
heart_pipeline = Pipeline([
    ('encoder', OrdinalEncoder())
])
```

NeverMarried:Ordinally encode.

```python
married_pipeline = Pipeline([
    ('encoder',OrdinalEncoder())
])
```

Occupation: One hot encode.

Occupation_Govt_job

Occupation_Never_worked

Occupation_Private

Occupation_Self-employed

Occupation_children

```python
from sklearn.preprocessing import OneHotEncoder
occupation_pipeline = Pipeline([
    ('onehot', OneHotEncoder(
        categories=[['Govt_job', 'Never_worked', 'Private', 'Self-employed', 'children']],
        drop = None,
        sparse_output = False,
        handle_unknown = 'ignore'
    ))
])
```

LivesIn: Impute with most frequent then ordinally encode.

```python
livesin_pipeline = Pipeline([
    ('encoder',OrdinalEncoder())
])
```

GlucoseLevel: Impute with mean, then min-max scaling.

```python
from sklearn.preprocessing import MinMaxScaler
glucose_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy = 'mean')),
    ('scaler',MinMaxScaler())
])
```

BMI: Impute with mean, then standard scale.

```python
bmi_pipeline = Pipeline([
    ('impute', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])
```

SmokingStatus: Impute with most frequent, then one hot encode

SmokingStatus_formerly smoked

SmokingStatus_never smoked

SmokingStatus_smokes




```python
smoking_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(
        categories=[['formerly smoked', 'never smoked', 'smokes']],
        drop = None,
        sparse_output = False,
        handle_unknown = 'ignore'
    ))
])
```

```python
from sklearn.compose import ColumnTransformer
preprocessor = ColumnTransformer(transformers= [
    ('gender', gender_pipeline, ['Gender']),
    ('age', age_pipeline, ['Age']),
    ('hastension', tension_pipeline, ['HasTension']),
    ('anyheartdisease', heart_pipeline, ['AnyHeartDisease']),
    ('nevermarried', married_pipeline, ['NeverMarried']),
    ('occupation', occupation_pipeline, ['Occupation']),
    ('livesin', livesin_pipeline, ['LivesIn']),
    ('glucoselevel', glucose_pipeline, ['GlucoseLevel']),
    ('bmi', bmi_pipeline, ['BMI']),
    ('smoking', smoking_pipeline, ['SmokingStatus'])
])
```

```python
features_train_processed = preprocessor.fit_transform(features_train)
```
